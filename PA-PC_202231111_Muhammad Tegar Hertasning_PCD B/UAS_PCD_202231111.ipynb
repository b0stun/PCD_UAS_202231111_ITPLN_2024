```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path = 'TipeXX.jpeg'  
original_image = cv2.imread(image_path)

def show_image(img, title):
    plt.figure(figsize=(5, 5))
    if len(img.shape) == 3:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.title(title)
    plt.axis('off')
    plt.show()

show_image(original_image, 'Citra Asli / Original image')
```

```python
(h, w) = original_image.shape[:2]
center = (w / 2, h / 2)
M = cv2.getRotationMatrix2D(center, 60, 1.0)
rotated_image = cv2.warpAffine(original_image, M, (w, h))
show_image(rotated_image, 'Rotated')
```

```python
resized_image = cv2.resize(original_image, (w // 2, h // 1))

show_image(resized_image, 'Resized Image')
```

```python
start_x = w // 4
start_y = h // 2
end_x = start_x + w // 2
end_y = start_y + h // 2
cropped_image = original_image[start_y:end_y, start_x:end_x]

show_image(cropped_image, 'Cropped Image')
```

```python
flipped_image = cv2.flip(original_image, 1)

show_image(flipped_image, 'Flipped Image')
```

```python
M = np.float32([[1, 0, 60], [0, 1, 40]])
translated_image = cv2.warpAffine(original_image, M, (w, h))
show_image(translated_image, 'Translated Image')
```
